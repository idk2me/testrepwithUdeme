```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path = r'C:\webdrivers\chromedriver.exe')
```

```python
brands = [] #List to store brand ofthe product
ratings = [] #List to store rating of the product
products = [] #List to store name of the product
promos = [] #List to store promo offers
prices = [] #List to store price of the product
discounted_prices = [] #List to store discounted price of the product
offers = [] #List to store number of offers of the product
shipping_cost = [] #List to store shipping charges of the product
driver.get("https://www.newegg.com/Video-Cards-Video-Devices/Category/ID-38?Tpk=graphics%20cards")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
containers = soup.findAll('div', attrs = {'class': 'item-info'})
print(len(containers))
print(BeautifulSoup.prettify(containers[0]))
```

```python
for container in containers:
    brand = container.find('img')['title']
    brands.append(brand)
    rating = container.find('a', href = True, attrs = {'class': 'item-rating'})
    rating = rating['title'] if rating is not None else ""
    ratings.append(rating)
    #print(rating)
    product = container.find('a', href = True, attrs = {'class': 'item-title'}).text
    products.append(product)
    promo = container.find('p', attrs = {'class': 'item-promo'}).text
    promos.append(promo)
    #print(product + promo)
    amount = container.find('div', attrs = {'class': 'item-action'})
    price = amount.strong.text + amount.sup.text
    prices.append(price)
    discounted_price = container.find('span', attrs = {'class': 'price-note-dollar'})
    #print(discounted_price)
    discounted_price = discounted_price['data-price'] if discounted_price is not None else ''
    discounted_prices.append(discounted_price)
    offer = container.find('a', href = True, attrs = {'class': 'price-current-num'})
    offer = offer.text if offer is not None else ""
    offers.append(offer)
    shipping = container.find('li', attrs = {'class': 'price-ship'}).text.strip()
    shipping_cost.append(shipping)

# print(len(brands))
# print(len(discounted_prices))
# print(len(ratings))
# print(len(offers))
# print(len(products))
```

```python
df = pd.DataFrame({'Brand Name':brands, 'Ratings':ratings, 'Product Name':products,
                   'Valid Promos':promos, 'Price':prices, 'Discounted Price':discounted_price,
                   'Offers':offers, 'Shipping Charges':shipping_cost,})

df.head(20)
```
